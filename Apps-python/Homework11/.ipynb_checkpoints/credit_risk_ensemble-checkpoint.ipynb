## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

```python
# Integer Encoding - get_dummies()
df_encoded = pd.get_dummies(df, columns=["home_ownership", "verification_status", "loan_status", "pymnt_plan", "initial_list_status", "application_type", "hardship_flag", "debt_settlement_flag"])
df_encoded.head()
```

```python
# OR Integer Encoding - LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# home_ownership column
le.fit(df["home_ownership"])
df["home_ownership"] = le.transform(df["home_ownership"])

# verification_status column
le.fit(df["verification_status"])
df["verification_status"] = le.transform(df["verification_status"])

# loan_status column
le.fit(df["loan_status"])
df["loan_status"] = le.transform(df["loan_status"])

# pymnt_plan column
le.fit(df["pymnt_plan"])
df["pymnt_plan"] = le.transform(df["pymnt_plan"])

# initial_list_status column
le.fit(df["initial_list_status"])
df["initial_list_status"] = le.transform(df["initial_list_status"])

# application_type column
le.fit(df["application_type"])
df["application_type"] = le.transform(df["application_type"])

# hardship_flag column
le.fit(df["hardship_flag"])
df["hardship_flag"] = le.transform(df["hardship_flag"])

# debt_settlement_flag column
le.fit(df["debt_settlement_flag"])
df["debt_settlement_flag"] = le.transform(df["debt_settlement_flag"])

# next_pymnt_d column
le.fit(df["next_pymnt_d"])
df["next_pymnt_d"] = le.transform(df["next_pymnt_d"])

# issue_d column
le.fit(df["issue_d"])
df["issue_d"] = le.transform(df["issue_d"])

df.head()
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
brf = BalancedRandomForestClassifier(n_estimators=1000, random_state=1)
brf_model = brf.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
y_pred = brf_model.predict(X_test_scaled)
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = brf_model.feature_importances_
```

```python
# List the features sorted in descending order by feature importance
sorted(zip(brf_model.feature_importances_, X.columns), reverse=True)
```

```python
# Visualize the features by importance
importances_df = pd.DataFrame(sorted(zip(brf_model.feature_importances_, X.columns), reverse=True))
importances_df.set_index(importances_df[1], inplace=True)
importances_df.drop(columns=1, inplace=True)
importances_df.rename(columns={0: 'Feature Importances'}, inplace=True)
importances_sorted = importances_df.sort_values(by='Feature Importances')
importances_sorted.plot(kind='barh', color='lightgreen', title= 'Features Importances', legend=False)
```
