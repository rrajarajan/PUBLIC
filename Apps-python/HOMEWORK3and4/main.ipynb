```python
#### INITIAL IMPORTS ####
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline

#### PREPARE AN CLEAN DATA ####
# Read WHALE_RETURNS
whale_data = Path("./Resources/whale_returns.csv")
whale_df = pd.read_csv(whale_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_df = whale_df.sort_index()

# Count nulls - whale_returns
whale_df_null = whale_df.isnull().count()
whale_df_null
```

```python
#Drop NULL - whale_returns
whale_df = whale_df.dropna().copy()
whale_df.head()
```

```python
# PLOT -  - whale_returns
whale_df.plot(figsize=(20,10), title="WHALE DATA")
```

```python
# Read ALGO_RETURNS
algo_data = Path("./Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_df = algo_df.sort_index()

# Count nulls - Algo_returns
algo_df_null = algo_df.isnull().count()
algo_df_null 
```

```python
#Drop NULL - Algo_returns
algo_df = algo_df.dropna().copy()
algo_df.head()
```

```python
# PLOT algo_returns
algo_df.plot(figsize=(20,10), title="ALGO DATA")
```

```python
# Read SP_TSX HISTORY
sp_tsx_data = Path("./Resources/sp_tsx_history.csv")
sp_tsx_df = pd.read_csv(sp_tsx_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp_tsx_df = sp_tsx_df.sort_index()
sp_tsx_df.head()
```

```python
# Check Data Types
sp_tsx_df.dtypes
```

```python
# Fix Data Types
#clean and remove $ and ,
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace("$","")
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace(",","")
sp_tsx_df.head()
```

```python
sp_tsx_df["Close"]
# Convert `order_total` from `object` to `float`
sp_tsx_df["Close"] = sp_tsx_df["Close"].astype("float")
sp_tsx_df.dtypes
```

```python
# PLOT - SP TSX
sp_tsx_df.plot(figsize=(20,10), title="SP TSX DATA")
```

```python
#CALCULATE Daily Daily Returns - SP TSX
daily_returns_method1 = (sp_tsx_df - sp_tsx_df.shift(1)) / sp_tsx_df.shift(1)
daily_returns_method1.head()

# Calculate the daily return using the 'pct_change()'
sp_tsx_dr = sp_tsx_df.pct_change()

#Drop NULL
sp_tsx_dr = sp_tsx_dr.dropna().copy()  
sp_tsx_dr.head()    
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_dr = sp_tsx_dr.rename(columns={"Close":"SP TSX"})
sp_tsx_dr.head()
```

```python
# Plot the daily returns of the SP_TSX over the last 5 years
sp_tsx_dr.plot(figsize=(20,10), title="SP TSX - DAILY RETURNS")
```

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
# concat/ JOIN All the Daily returns
joined_dr = pd.concat([whale_df,algo_df,sp_tsx_dr], axis="columns", join="inner")
joined_dr = joined_dr.rename(columns={
    "Close": "SP TSX"
})
joined_dr.head()
```

```python
#### Conduct Quantitative Analysis ####
# Plot daily returns of all portfolios
joined_dr.plot(figsize=(20,10), title="DAILY RETURNS OF ALL PORTFOLIOS")
```

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1 + joined_dr).cumprod()
cumulative_returns.head()
```

```python
#Plot cumulative returns
cumulative_returns.plot(figsize=(20,10),  title="CUMULATIVE RETURNS")
```

```python
### RISK ANALYSIS #####
# Box plot to visually show risk
joined_dr.plot.box(figsize=(20,10), title="BOXPLOT - PORTFOLIO")
```

```python
# CALCULATE - Daily Standard Deviations
joined_dr_std = joined_dr.std()
joined_dr_std.head()
```

```python
# Determine which portfolios are riskier than the S&P TSX 60
# Sort std in desc order
joined_dr_std = joined_dr_std.sort_values(ascending=False)
joined_dr_std.head()
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_daily_returns_std = joined_dr_std * np.sqrt(252)
annualized_daily_returns_std.head()
```

```python
#### ROLLING STATISTICS ########
joined_dr.head()

# Calculate the rolling standard deviation for all portfolios using a 21-day window
joined_dr.rolling(window=21).std().plot(figsize=(20,10), title="ROLLING MEAN")
```

```python
# Calculate the MEAN for all portfolios using a 21-day window
joined_dr.rolling(window=21).mean().plot(figsize=(20,10), title="ROLLING MEAN")
```

```python
#### CORRELATION ########
# Calculate the correlation
correlation = joined_dr.corr()
correlation
```

```python
# Display de correlation matrix
# Import the Seaborn library
import seaborn as sns
sns.heatmap(correlation, vmin=-1, vmax=1)
```

```python
#### BETA, CONVARIANCE ####
joined_dr.head()
# Calculate variance of S&P TSX
variance = joined_dr['SP TSX'].var()
variance
```

```python
# Calculate covariance of a single portfolio
#ALGO 1
#covariance
covariance = joined_dr['Algo 1'].cov(joined_dr['SP TSX'])
covariance
```

```python
# Computing beta
algo1_beta = covariance / variance
algo1_beta
```

```python
# Plot beta trend
rolling_covariance = joined_dr['Algo 1'].rolling(window=30).cov(joined_dr['SP TSX'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 30-Day Covariance of ALGO 1 Returns vs. SP TSX Returns')
```

```python
# Calculate covariance of a single portfolio
#ALGO 2
#covariance
covariance2 = joined_dr['Algo 2'].cov(joined_dr['SP TSX'])
covariance2
```

```python
# Computing beta
algo2_beta = covariance2 / variance
algo2_beta
```

```python
# Plot beta trend
rolling_covariance = joined_dr['Algo 2'].rolling(window=30).cov(joined_dr['SP TSX'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 30-Day Covariance of ALGO 2 Returns vs. SP TSX Returns')
```

```python
 # Showcase beta vs. correlation by plotting a scatterplot using the Seaborn library and fitting a regression line
import seaborn as sns

sns.lmplot(x="SP TSX", y="Algo 1", data=joined_dr, aspect=1.5, fit_reg=True)
```

```python
sns.lmplot(x="SP TSX", y="Algo 2", data=joined_dr, aspect=1.5, fit_reg=True)
```

```python
#### SHARPE RATIO #####
# get joined DAILY RETURN
joined_dr.head()

# Calculate Sharpe Ratio
sharpe_ratios = (joined_dr.mean() * 252) / (joined_dr_std.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Plot sharpe ratios
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
#Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.
sharpe_ratios = sharpe_ratios.sort_values(ascending=False)
sharpe_ratios
```
