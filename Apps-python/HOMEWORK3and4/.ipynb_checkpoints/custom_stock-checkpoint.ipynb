```python
### PREPARE AND CLEAN DATA ### 
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Reading data from 1st stock
# Read SHOPIFY
shop_df = Path("./Resources/shop_historical.csv")
shop_df = pd.read_csv(shop_df, index_col="Date", infer_datetime_format=True, parse_dates=True)
shop_df = shop_df.rename(columns={"Close":"SHOP"})
shop_df = shop_df.drop(columns=["Symbol"])
shop_df = shop_df.sort_index()
shop_df.head()
```

```python
#Plot Shopify
shop_df.plot(figsize=(20,10), title="SHOPIFY DATA")
```

```python
# Reading data from 2nd stock
# Read OTEX
otex_df = Path("./Resources/otex_historical.csv")
otex_df = pd.read_csv(otex_df, index_col="Date", infer_datetime_format=True, parse_dates=True)
otex_df = otex_df.rename(columns={"Close":"OTEX"})
otex_df = otex_df.drop(columns=["Symbol"])
otex_df = otex_df.sort_index()
otex_df.head()
```

```python
#Plot OTEX
otex_df.plot(figsize=(20,10), title="OTEX DATA")
```

```python
# Reading data from 3rd stock
# Read L
l_df = Path("./Resources/l_historical.csv")
l_df = pd.read_csv(l_df, index_col="Date", infer_datetime_format=True, parse_dates=True)
l_df = l_df.rename(columns={"Close":"L"})
l_df = l_df.drop(columns=["Symbol"])
l_df = l_df.sort_index()
l_df.head()
```

```python
#Plot L
l_df.plot(figsize=(20,10), title="L DATA")
```

```python
# Create a new pivot table where the columns are the closing prices for each ticker
all_prices = pd.concat([shop_df, otex_df,l_df], axis="columns", join="inner")
# Sort datetime index in ascending order (past to present)
all_prices = all_prices.sort_index()
all_prices.head()
```

```python
all_prices.dtypes
```

```python
# Calculate Daily Returns
all_returns = all_prices.pct_change()

#Drop NULL
all_returns = all_returns.dropna().copy()  
all_returns.head()  

all_returns.head()
```

```python
# Use the `std` function and multiply by the square root of the number of trading days in a year to get annualized volatility
volatility = all_returns.std() * np.sqrt(252)

# Sort values in ascending order
volatility = volatility.sort_values()

# Display the values
volatility
```

```python
# Shortcut for calculating returns
weights = [0.33,0.33,0.33]
portfolio_returns_short = all_returns.dot(weights)
portfolio_returns_short.head()
```

```python
#### Conduct Quantitative Analysis ####
# Plot daily returns of all portfolios
all_returns.plot(figsize=(20,10), title="COMBINED - DAILY RETURNS")
```

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1 + all_returns).cumprod()
cumulative_returns.head()
```

```python
#Plot cumulative returns
cumulative_returns.plot(figsize=(20,10),  title="CUMULATIVE RETURNS")
```

```python
# Box plot to visually show risk
all_returns.plot.box(figsize=(20,10), title="BOXPLOT - PORTFOLIO")
```

```python
# Daily Standard Deviations
all_returns_std = all_returns.std()
all_returns_std = all_returns_std.sort_values(ascending=False)
all_returns_std.head()
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_all_returns_std = all_returns_std * np.sqrt(252)
annualized_all_returns_std.head()
```

```python
#### ROLLING STATISTICS ########
all_returns.head()

# Calculate the rolling standard deviation for all portfolios using a 21-day window
all_returns.rolling(window=21).mean().plot(figsize=(20,10), title="ROLLING MEAN")
```

```python
#### CORRELATION ########
# Calculate the correlation
correlation = all_returns.corr()
correlation
```

```python
# Display de correlation matrix
# Import the Seaborn library
import seaborn as sns
sns.heatmap(correlation, vmin=-1, vmax=1)
```

```python
### GET, PREPARE and CLEAN DATA SP_TSX ###
# Read the SP_TSX HISTORY
sp_tsx_data = Path("./Resources/sp_tsx_history.csv")
sp_tsx_df = pd.read_csv(sp_tsx_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp_tsx_df = sp_tsx_df.rename(columns={"Close":"SP TSX"})
sp_tsx_df = sp_tsx_df.sort_index()
sp_tsx_df.head()
```

```python
# Check Data Types
sp_tsx_df.dtypes
```

```python
# Fix Data Types
#clean and remove $ and ,
sp_tsx_df["SP TSX"] = sp_tsx_df["SP TSX"].str.replace("$","")
sp_tsx_df["SP TSX"] = sp_tsx_df["SP TSX"].str.replace(",","")
sp_tsx_df.head()
```

```python
sp_tsx_df["SP TSX"]
# Convert `order_total` from `object` to `float`
sp_tsx_df["SP TSX"] = sp_tsx_df["SP TSX"].astype("float")
sp_tsx_df.dtypes
```

```python
# PLOT SP TSX
sp_tsx_df.plot(figsize=(20,10), title="SP TSX DATA")
```

```python
# Calculate the daily return using the 'pct_change()'
sp_tsx_dr = sp_tsx_df.pct_change()
#Drop NULL
sp_tsx_dr = sp_tsx_dr.dropna().copy()  
sp_tsx_dr.head()    
```

```python
# Plot the daily returns of the SP_TSX over the last 5 years
sp_tsx_dr.plot(figsize=(20,10), title="SP TSX - DAILY RETURNS")
```

```python
# Join all_returns and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
# concat/ JOIN All the Daily returns
joined_dr = pd.concat([all_returns,sp_tsx_dr], axis="columns", join="inner")
joined_dr.head()
```

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1 + joined_dr).cumprod()
cumulative_returns.head()
```

```python
#Plot cumulative returns
cumulative_returns.plot(figsize=(20,10),  title="CUMULATIVE RETURNS")
```

```python
### RISK ANALYSIS #####
# Box plot to visually show risk
joined_dr.plot.box(figsize=(20,10), title="BOXPLOT - PORTFOLIO")
```

```python
#### BETA, CONVARIANCE ####
joined_dr.head()
# Calculate variance of S&P TSX
variance = joined_dr['SP TSX'].var()
variance
```

```python
# Calculate covariance of a single portfolio
#SHOP
#covariance
covariance = joined_dr['SHOP'].cov(joined_dr['SP TSX'])
covariance
```

```python
# Computing beta
shop_beta = covariance / variance
shop_beta
```

```python
# Plot beta trend
rolling_covariance = joined_dr['SHOP'].rolling(window=30).cov(joined_dr['SP TSX'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 30-Day Covariance of SHOP Returns vs. SP TSX Returns')
```

```python
import seaborn as sns

#PLOT SCATTER GRAPH
sns.lmplot(x="SP TSX", y="SHOP", data=joined_dr, aspect=1.5, fit_reg=True)
```

```python
#### SHARPE RATIO #####
# get joined DAILY RETURN
joined_dr.head()

# Calculate Sharpe Ratio
sharpe_ratios = (joined_dr.mean() * 252) / (joined_dr.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Plot sharpe ratios
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
sharpe_ratios = sharpe_ratios.sort_values(ascending=False)
sharpe_ratios
```
